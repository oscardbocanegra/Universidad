```python
import unicodedata

    # Normalizar texto: quitar acentos, pasar a mayúsculas y limpiar espacios/guiones
def _normalize_text(s):
    if pd.isna(s):
        return s
    s = str(s).strip().upper()
    s = ''.join(c for c in unicodedata.normalize('NFKD', s) if not unicodedata.combining(c))
    s = s.replace('-', ' ').replace('_', ' ')
    s = " ".join(s.split())
    return s
    # Ver valores actuales únicos (por si quieres revisar antes de unificar)
print("Valores únicos originales de ENTIDAD:")
print(df['ENTIDAD'].unique())

    # Aplicar normalización básica
df['ENTIDAD_norm'] = df['ENTIDAD'].apply(_normalize_text)

    # Mapeo manual de variantes a nombres unificados (ajusta según los valores reales)
mapping_entidades = {
    'INVIAS OTROS': 'INVIAS',
    'INVIAS-OTROS': 'INVIAS',
    'INVIAS': 'INVIAS',
    'ANI': 'ANI',
    'ANi': 'ANI',  # ejemplo si hubiera variaciones con minúsculas o errores
        # agrega aquí más reglas si detectas otras variantes
}

    # Aplicar mapeo; si no hay mapeo, dejar el valor normalizado
df['ENTIDAD_unificada'] = df['ENTIDAD_norm'].replace(mapping_entidades).fillna(df['ENTIDAD_norm'])

    # (Opcional) Reemplazar la columna original ENTIDAD por la versión unificada
df['ENTIDAD'] = df['ENTIDAD_unificada']

    # Mostrar resumen antes/después
print("\nConteo por ENTIDAD unificada:")
print(df['ENTIDAD'].value_counts())

    # Eliminar columnas intermedias si no se necesitan
df.drop(columns=['ENTIDAD_norm', 'ENTIDAD_unificada'], inplace=True)
```
