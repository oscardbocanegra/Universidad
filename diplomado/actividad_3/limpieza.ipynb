```python
# Lista "canónica" en minúsculas (nombres de campo API)
cols_to_drop = {
    # 1) Identificadores / PII
    "estu_tipodocumento", "estu_consecutivo", "estu_fechanacimiento",
    "estu_genero", "estu_nacionalidad", "estu_pais_reside",
    "estu_estudiante", "estu_estadoinvestigacion", "estu_privado_libertad",

    # 2) Identificación del plantel
    "cole_nombre_establecimiento", "cole_nombre_sede",
    "cole_codigo_icfes", "cole_cod_dane_establecimiento", "cole_cod_dane_sede",

    # 3) Atributos del plantel no usados
    "cole_area_ubicacion", "cole_bilingue", "cole_calendario",
    "cole_caracter", "cole_genero", "cole_jornada", "cole_sede_principal",

    # 4) Geografía duplicada (dejamos presentación por nombre)
    "cole_mcpio_ubicacion", "cole_depto_ubicacion",
    "cole_cod_depto_ubicacion", "cole_cod_mcpio_ubicacion",
    "estu_depto_reside", "estu_mcpio_reside",
    "estu_cod_reside_depto", "estu_cod_reside_mcpio",
    "estu_cod_depto_presentacion", "estu_cod_mcpio_presentacion",

    # 5) Socioeconómicas no usadas (dejamos solo fami_estratovivienda)
    "fami_cuartoshogar", "fami_personashogar",
    "fami_tieneautomovil", "fami_tienecomputador",
    "fami_tieneinternet", "fami_tienelavadora",
    "fami_educacionmadre", "fami_educacionpadre",

    # 6) Puntajes por área (usamos punt_global)
    "desemp_ingles", "punt_ingles", "punt_matematicas",
    "punt_sociales_ciudadanas", "punt_c_naturales", "punt_lectura_critica",
}

# Buscar cuáles de esas columnas existen en df (independiente de mayúsculas/minúsculas)
lower_map = {c.lower(): c for c in df.columns}  # mapa lower -> nombre real
present_cols = [lower_map[c] for c in cols_to_drop if c in lower_map]
```

# Data con columnas eliminadas pasa a df_clean

```python
df_clean = df.drop(columns=present_cols)


print(f"Columnas eliminadas ({len(present_cols)}): {present_cols}")
```

```python
df_clean.shape
```

```python
# Helpers para encontrar las columnas aunque cambien mayúsculas/minúsculas
def find_col(df, name):
    name = name.lower()
    for c in df.columns:
        if c.lower() == name:
            return c
    return None

score_col   = find_col(df_clean, "punt_global")
nat_col     = find_col(df_clean, "cole_naturaleza")
estrato_col = find_col(df_clean, "fami_estratovivienda")

# --- 1) Boxplot global (puntaje) ---
serie = pd.to_numeric(df_clean[score_col], errors="coerce").dropna()

plt.figure()
plt.boxplot(serie.values)
plt.title("Outliers del Puntaje Global")
plt.ylabel("Puntaje")
plt.xlabel("Distribución")
plt.show()

# (Opcional) Conteo de outliers por regla 1.5*IQR
Q1, Q3 = serie.quantile(0.25), serie.quantile(0.75)
IQR = Q3 - Q1
outliers = serie[(serie < Q1 - 1.5*IQR) | (serie > Q3 + 1.5*IQR)]
print(f"Outliers detectados (global): {outliers.size}")

# --- 2) Boxplot por naturaleza del colegio ---
if nat_col:
    data, labels = [], []
    for g in sorted(df_clean[nat_col].dropna().unique()):
        vals = pd.to_numeric(df_clean.loc[df_clean[nat_col] == g, score_col], errors="coerce").dropna().values
        if len(vals) > 0:
            data.append(vals); labels.append(str(g))

    if data:
        plt.figure()
        plt.boxplot(data, labels=labels)
        plt.title("Puntaje Global por Naturaleza del Colegio")
        plt.ylabel("Puntaje")
        plt.xlabel("Naturaleza del colegio")
        plt.show()

# --- 3) Boxplot por estrato de vivienda ---
if estrato_col:
    # Ordena los estratos por su número (si existiera texto, lo ignora)
    uniq = df_clean[estrato_col].dropna().astype(str).unique()
    order = sorted(uniq, key=lambda x: int("".join(ch for ch in x if ch.isdigit()) or 99))

    data, labels = [], []
    for g in order:
        vals = pd.to_numeric(df_clean.loc[df_clean[estrato_col].astype(str) == g, score_col], errors="coerce").dropna().values
        if len(vals) > 0:
            data.append(vals); labels.append(g)

    if data:
        plt.figure()
        plt.boxplot(data, labels=labels)
        plt.title("Puntaje Global por Estrato de Vivienda")
        plt.ylabel("Puntaje")
        plt.xlabel("Estrato")
        plt.show()
```

```python
import numpy as np

# --- Helpers ---
def find_col(df, name):
    name = name.lower()
    for c in df.columns:
        if c.lower() == name:
            return c
    raise KeyError(f"No encontré la columna '{name}' en df_clean.columns")

score_col = find_col(df_clean, "punt_global")
nat_col   = find_col(df_clean, "cole_naturaleza")

# --- Preparación ---
tmp = df_clean[[score_col, nat_col]].copy()
tmp[score_col] = pd.to_numeric(tmp[score_col], errors="coerce")
tmp = tmp.dropna(subset=[score_col, nat_col])

# Normaliza categorías y crea etiquetas legibles
canon = (
    tmp[nat_col].astype(str).str.strip().str.lower()
            .str.replace(r"[_\-]+", " ", regex=True)
)

label_map = {
    "oficial": "Colegios Públicos (Oficial)",
    "no oficial": "Colegios Privados (No Oficial)",
    "privado": "Colegios Privados (No Oficial)"
}
tmp["__canon__"] = canon
tmp["__label__"] = tmp["__canon__"].map(label_map).fillna(tmp[nat_col].astype(str))

# --- Agregación: media, n, std, IC95% ---
g = (
    tmp.groupby(["__canon__", "__label__"])[score_col]
       .agg(mean="mean", count="count", std="std")
       .reset_index()
)
g["sem"] = g["std"] / np.sqrt(g["count"].clip(lower=1))
g["ci95"] = 1.96 * g["sem"]

# Orden preferido: Oficial, No Oficial (si existen)
order = []
if "oficial" in g["__canon__"].values: order.append("oficial")
if "no oficial" in g["__canon__"].values: order.append("no oficial")
# agrega cualquier otra categoría al final si aparece
order += [c for c in g["__canon__"].unique() if c not in order]

g = g.set_index("__canon__").loc[order].reset_index()

# --- Gráfica ---
plt.figure()
plt.bar(range(len(g)), g["mean"].values, yerr=g["ci95"].values, capsize=4)
plt.xticks(range(len(g)), g["__label__"].tolist(), rotation=0)
plt.ylabel("Puntaje Global Promedio")
plt.xlabel("Naturaleza del Colegio")
plt.title("Gráfica 1: Brecha de Rendimiento (Puntaje Promedio)")
plt.grid(axis="y", linestyle="--", alpha=0.4)

# Anota la diferencia entre Oficial y No Oficial (si ambas existen)
if {"oficial", "no oficial"}.issubset(set(g["__canon__"])):
    m_of = g.loc[g["__canon__"] == "oficial", "mean"].iloc[0]
    m_no = g.loc[g["__canon__"] == "no oficial", "mean"].iloc[0]
    diff = m_no - m_of
    ymax = (g["mean"] + g["ci95"]).max()
    plt.text(0.5, ymax * 1.02, f"Diferencia: {diff:.1f} pts",
             ha="center", va="bottom", fontsize=10)

plt.tight_layout()
plt.show()
```

```python
def find_col(df, name):
    name = name.lower()
    for c in df.columns:
        if c.lower() == name:
            return c
    raise KeyError(f"No encontré la columna '{name}' en el DataFrame.")

score_col = find_col(df_clean, "punt_global")
nat_col   = find_col(df_clean, "cole_naturaleza")

# Prepara datos
tmp = df_clean[[score_col, nat_col]].copy()
tmp[score_col] = pd.to_numeric(tmp[score_col], errors="coerce")
tmp = tmp.dropna(subset=[score_col, nat_col])

# Normaliza y mapea etiquetas legibles
canon = (
    tmp[nat_col].astype(str).str.strip().str.lower()
       .str.replace(r"[_\-]+", " ", regex=True)
)
label_map = {
    "oficial": "Colegios Públicos (Oficial)",
    "no oficial": "Colegios Privados (No Oficial)",
    "privado": "Colegios Privados (No Oficial)",
}
tmp["__canon__"] = canon
tmp["__label__"] = tmp["__canon__"].map(label_map).fillna(tmp[nat_col].astype(str))

# Orden preferido: Oficial, No Oficial, otros al final si existieran
order = []
if "oficial" in tmp["__canon__"].values: order.append("oficial")
if "no oficial" in tmp["__canon__"].values: order.append("no oficial")
order += [c for c in tmp["__canon__"].unique() if c not in order]

# Datos para el boxplot
data, labels = [], []
for cat in order:
    vals = tmp.loc[tmp["__canon__"] == cat, score_col].dropna().values
    if len(vals) > 0:
        data.append(vals)
        # toma etiqueta legible (si existe) de la primera fila del grupo
        lab = tmp.loc[tmp["__canon__"] == cat, "__label__"].iloc[0]
        labels.append(lab)

# Gráfica
plt.figure()
plt.suptitle("Gráfica 2 (Boxplot)")
plt.boxplot(data, labels=labels)
plt.ylabel("Puntaje Global")
plt.xlabel("Naturaleza del Colegio")
plt.title("Gráfica 2: Distribución de Puntajes (Box Plot)")
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```
