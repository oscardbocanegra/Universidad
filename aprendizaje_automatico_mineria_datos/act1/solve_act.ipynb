Determinar el conjunto de modelización y el de validación.

```python
from sklearn.model_selection import train_test_split

# Dividir el dataframe df_day en conjuntos de entrenamiento y validación
train_df_day, val_df_day = train_test_split(df_day, test_size=0.2, random_state=42)

# Dividir el dataframe df2_hour en conjuntos de entrenamiento y validación
train_df2_hour, val_df2_hour = train_test_split(df2_hour, test_size=0.2, random_state=42)

# Mostrar el tamaño de los conjuntos
print(f"Tamaño del conjunto de entrenamiento (df_day): {train_df_day.shape}")
print(f"Tamaño del conjunto de validación (df_day): {val_df_day.shape}")
print(f"Tamaño del conjunto de entrenamiento (df2_hour): {train_df2_hour.shape}")
print(f"Tamaño del conjunto de validación (df2_hour): {val_df2_hour.shape}")
```

	Correlaciones. 

```python
# Select only numeric columns
numeric_df = df_day.select_dtypes(include=['number'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

```

```python
# Select only numeric columns
numeric_df = df2_hour.select_dtypes(include=['number'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()

```

	Distribuciones

```python
# Visualización de la distribución de la variable cnt
plt.figure(figsize=(12, 5))

# Distribución en datos diarios
plt.subplot(1, 2, 1)
sns.histplot(df_day["cnt"], bins=30, kde=True, color="blue")
plt.title("Distribución de alquileres diarios (cnt)")
plt.xlabel("Cantidad de alquileres")

# Distribución en datos horarios
plt.subplot(1, 2, 2)
sns.histplot(df2_hour["cnt"], bins=30, kde=True, color="red")
plt.title("Distribución de alquileres por hora (cnt)")
plt.xlabel("Cantidad de alquileres")

plt.tight_layout()
plt.show()
```

	Gráficos que consideres adecuados.

```python
# Gráfico de alquileres por estación del año
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_day["season"], y=df_day["cnt"], hue=df_day["season"], palette="coolwarm", legend=False)
plt.title("Distribución de alquileres por estación del año")
plt.xlabel("Estación (1: Primavera, 2: Verano, 3: Otoño, 4: Invierno)")
plt.ylabel("Cantidad de alquileres")
plt.show()

# Gráfico de alquileres por hora del día (solo en hour.csv)
plt.figure(figsize=(10, 5))
sns.boxplot(x=df2_hour["hr"], y=df2_hour["cnt"], hue=df2_hour["hr"], palette="coolwarm", legend=False)
plt.title("Distribución de alquileres por hora del día")
plt.xlabel("Hora del día")
plt.ylabel("Cantidad de alquileres")
plt.show()
```

```python
# Convertir la columna de fecha a tipo datetime para mejorar la visualización
df_day["dteday"] = pd.to_datetime(df_day["dteday"])

# Graficar nuevamente con mejor formato de fechas
plt.figure(figsize=(12, 5))
sns.lineplot(x=df_day["dteday"], y=df_day["cnt"], color="blue")

# Mejorar la visualización de las fechas en el eje X
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(10))  # Mostrar un número limitado de fechas

plt.title("Evolución de alquileres diarios a lo largo del tiempo")
plt.xlabel("Fecha")
plt.ylabel("Cantidad de alquileres")
plt.show()
```
