	Correlaciones. 

````python
# Convertir la columna 'dteday' a formato de fecha en df_day
df_day['dteday'] = pd.to_datetime(df_day['dteday'])

# Calcular las correlaciones para df_day
correlation_df_day = df_day.corr()
print("Correlaciones del dataframe df_day:")
print(correlation_df_day)

````

	Distribuciones

````python
import seaborn as sns

import matplotlib.pyplot as plt

# Distribuciones para el dataframe df_day
plt.figure(figsize=(15, 10))
for i, column in enumerate(df_day.columns, 1):
    plt.subplot(4, 4, i)
    sns.histplot(df_day[column], kde=True)
    plt.title(f'Distribución de {column}')
plt.tight_layout()
plt.show()

# Distribuciones para el dataframe df2_hour
plt.figure(figsize=(15, 10))
for i, column in enumerate(df2_hour.columns, 1):
    plt.subplot(5, 4, i)
    sns.histplot(df2_hour[column], kde=True)
    plt.title(f'Distribución de {column}')
plt.tight_layout()
plt.show()
````

	Gráficos que consideres adecuados.

````python
# Gráfico de líneas para ver la tendencia de cnt a lo largo del tiempo
plt.figure(figsize=(14, 7))
plt.plot(df_day['dteday'], df_day['cnt'], label='Total Count')
plt.xlabel('Date')
plt.ylabel('Count')
plt.title('Trend of Total Count Over Time')
plt.legend()
plt.show()


# Gráfico de barras para comparar la cantidad de bicicletas alquiladas por estación
plt.figure(figsize=(10, 6))
sns.barplot(x='season', y='cnt', data=df_day, estimator=sum, errorbar=None)
plt.xlabel('Season')
plt.ylabel('Total Count')
plt.title('Total Count by Season')
plt.show()

# Gráfico de dispersión para ver la relación entre la temperatura y la cantidad de bicicletas alquiladas
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temp', y='cnt', data=df_day)
plt.xlabel('Temperature')
plt.ylabel('Count')
plt.title('Relationship Between Temperature and Total Count')
plt.show()
````

````python
Sí, los dos últimos gráficos hacen referencia a los dataframes `df_day` y `df2_hour`. A continuación, te explico cada uno de ellos:

1. **Gráfico de líneas para ver la tendencia de `cnt` a lo largo del tiempo**:
    ```python
    plt.figure(figsize=(14, 7))
    plt.plot(df_day['dteday'], df_day['cnt'], label='Total Count')
    plt.xlabel('Date')
    plt.ylabel('Count')
    plt.title('Trend of Total Count Over Time')
    plt.legend()
    plt.show()
    ```
    Este gráfico de líneas muestra la tendencia del total de bicicletas alquiladas (`cnt`) a lo largo del tiempo en el dataframe `df_day`. En el eje X se representan las fechas (`dteday`) y en el eje Y el total de bicicletas alquiladas (`cnt`). Este gráfico ayuda a visualizar cómo varía el número de alquileres de bicicletas a lo largo del tiempo.

2. **Gráfico de barras para comparar la cantidad de bicicletas alquiladas por estación**:
    ```python
    plt.figure(figsize=(10, 6))
    sns.barplot(x='season', y='cnt', data=df_day, estimator=sum, errorbar=None)
    plt.xlabel('Season')
    plt.ylabel('Total Count')
    plt.title('Total Count by Season')
    plt.show()
    ```
    Este gráfico de barras muestra la cantidad total de bicicletas alquiladas (`cnt`) por estación (`season`) en el dataframe `df_day`. En el eje X se representan las estaciones y en el eje Y el total de bicicletas alquiladas. Este gráfico permite comparar el número de alquileres de bicicletas en diferentes estaciones del año.

3. **Gráfico de dispersión para ver la relación entre la temperatura y la cantidad de bicicletas alquiladas**:
    ```python
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='temp', y='cnt', data=df_day)
    plt.xlabel('Temperature')
    plt.ylabel('Count')
    plt.title('Relationship Between Temperature and Total Count')
    plt.show()
    ```
    Este gráfico de dispersión muestra la relación entre la temperatura (`temp`) y el total de bicicletas alquiladas (`cnt`) en el dataframe `df_day`. En el eje X se representa la temperatura y en el eje Y el total de bicicletas alquiladas. Este gráfico ayuda a visualizar si existe alguna correlación entre la temperatura y el número de alquileres de bicicletas.

Estos gráficos proporcionan una visión general de cómo varían los alquileres de bicicletas en función del tiempo, la estación y la temperatura.
````
